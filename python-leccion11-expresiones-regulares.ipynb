````python
email = open("email.txt", "r").read()
````

````python
import re
````

 - 1. Extraer el email de la persona que envi贸 el email (os deber铆an salir 2).  Pista  Tendremos que hacer dos busquedas:
    
    
     - Una para extraer toda la informaci贸n del remitente. El remitente siempre empieza por From:
    
    ```python
        info_remitente = re.findall('From:.*',email)
    ``` 

     - Otra para extraer el email 煤nicamente.


````python
info_remitente = re.findall("From: .*", email)
````

````python
info_remitente
````

````python
direcciones_email = []

for remitente in info_remitente:

    direcciones_email.append(re.findall("\w+@\w+\.com", remitente))
````

````python
direcciones_email
````

 - 2. Extraer el nombre de la persona que envi贸 el email (os deber铆an salir 2).  Pista  Tendremos que hacer dos busquedas:

     - Podemos usar el resultado de la primera b煤squeda del ejercicio anterior para sacar el nombre.

     - Otra para extraer el nombre 煤nicamente.


````python
nombre_completo = []

for remitente in info_remitente:

    nombre_completo.append(re.findall("\w+\.\s\w+\s\w+", remitente)) #Cogemos str, punto, espacio, str, espacio, str
````

````python
nombre_completo
````

````python
nombre_de_pila = []

for nombre in info_remitente:

    nombre_de_pila.append(re.findall("\s\w+\s", nombre))
````

````python
nombre_de_pila
````

     
 - 3. El d铆a en el que se mand贸 el email (os deber铆an salir 2).
 Pista  De la misma forma que antes buscamos por From: primero, ahora lo tendremos que hacer con Date: .*


````python
info_fecha = re.findall("Date: .*", email)
````

````python
info_fecha
````

````python
fecha = []

for date in info_fecha:

    fecha.append(re.findall("\d{2}\s\w{3}\s\d{4}", date)) # Este Regex es: dos d铆gitos, espacio, tres letras, espacio, cuatro d铆gitos
````

````python
fecha
````

 - 4. El asunto del email (os deber铆an salir 2).
 Pista  Busca primero "Subject:.* y despu茅s busca el patr贸n para extraer el asunto.


````python
info_asunto = re.findall("Subject:.*", email)
````

````python
info_asunto
````

````python
asuntos = []

for subject in info_asunto:

    asuntos.append(re.findall("[A-Z]+\s[A-Z]+\s.*[A-Z]+\s.*", subject))
````

````python
asuntos
````

 - 5. Guarda todos los resultados en un diccionario.

````python
diccionario_email ={"Email": direcciones_email, "Nombre completo": nombre_completo, "Nombre de pila": nombre_de_pila, "Fecha": fecha, "Asunto": asuntos}

````
