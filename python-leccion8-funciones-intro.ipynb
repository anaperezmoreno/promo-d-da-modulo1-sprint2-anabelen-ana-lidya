 -2 El objetivo es comparar cada par de enteros de 2 listas, y devolver una nueva lista con el n칰mero mayor de la comapraci칩n..
        
      Probad la funci칩n para las siguientes listas: 
        
      ```python
        arr1 = [13, 64, 15, 17, 88]
        arr2 = [23, 14, 53, 17, 80]
        resultado = [23, 64, 53, 17, 88]
      ```

    游눠 Pista 游눠 En Python existe el m칠todo zip.


````python
arr1 = [13, 64, 15, 17, 88]
arr2 = [23, 14, 53, 17, 80]
````

````python
def ganador(arr1, arr2):
    resultado = []
    for tupla in zip(arr1, arr2):
        if tupla[0] >= tupla[1]:
            resultado.append(tupla[0])
        
        elif tupla[0] < tupla[1]:
            resultado.append(tupla[1])
    return resultado
````

````python
ganador(arr1, arr2)
````

Nuestro c칩digo no tiene que ser case sensitive, es decir, si hay una letra en may칰scula y en min칰scula cuenta como dos apariciones.
        
    Probad la funci칩n para los siguientes strings:

    ```python
        "din"      =>  "((("
        "recede"   =>  "()()()"
        "Success"  =>  ")())())"
        "(( @"     =>  "))((" 
        "Ocvl@GamFLAFkixkS" => "((()(()()))))(()("
    ```

````python
def parentesis(string):
    resultado = ""
    string = string.lower()
    for i in string:
        if string.count(i) > 1:
            resultado += ")"
        else:
            resultado += "("
    return resultado
````

````python
parentesis("Ocvl@GamFLAFkixkS")
````

````python
parentesis("din")
````

````python
parentesis("recede")
````

````python
parentesis("Success")
````
